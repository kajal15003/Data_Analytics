```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Privacy_survey.xlsx")
cnct= df[df.columns[0]], df[df.columns[5:21]]
df= pd.concat(cnct, axis=1)
df= pd.concat(cnct, axis=1)
df.columns=['ID','Expertise','Gender','Education','Privacy_Issue','Privacy_Leakage','Effect_PL','Risk_data','Risk_Phase','Hide_info','More_privacy','Pri_example','Privacy_premise','Regulation','Reg_help','Resposibility','Reg_effect']
df
```

```python
# Dict for converting categories to bar chart heights
cat_to_bar = {'I am a professional who has deep expertise related to video surveillance systems (system developer, engineer, vendor, etc.)':1, 'I am very familiar with video surveillance system and have used/am using them':2,'I am very familiar with video surveillance system but have not worked with or used them':3,'I have never heard of video surveillance system': 4}
#cat_to_bar1={'Very much': 3, 'Not sure': 4, 'Not at all': 5}
df[['Expertise']] = df[['Expertise']].replace(cat_to_bar)
#data[['Z']] = data[['Z']].replace(cat_to_bar1)
df
```

```python
grouped = df.groupby("Expertise") # change "Column Header Name" to the name of the column needed to categorise or group the rows in the dataframe, 
keys = grouped.groups.keys() #create a dictionary list of the each group unique varibles in the specifed column of the dataframe.   
print(keys) #a cheeky debug to check it's working
ax=df.groupby("Expertise")['Expertise'].count().plot(kind='pie',autopct='%0.2f%%',figsize=(5,5),legend=True,ylabel='',labeldistance=None,shadow=True)
ax.legend(bbox_to_anchor=(0.9, 1.1), loc='upper left')
```

```python
for key in keys: #looping through each key 
        splitdf = grouped.get_group(key) # creating a temporary dataframe with only the values of the current key. 
        splitdf.to_excel(str(key)+".xlsx", engine='xlsxwriter') #write the temporary dataframe called 'splitdf' to an excel file named after the key. At the end of the loop the temporary dataframe 'splitdf' is overwritten for use with the next key. 
```
